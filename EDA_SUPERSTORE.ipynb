```python
print("Top Selling Products: ")
top_selling_products = df.groupby('Sub-Category')['Sales'].sum().sort_values(ascending=False)
print(top_selling_products.head())
```

```python
print("Region making Hightest Loss :")
region_highest_loss = df.groupby('Region')['Profit'].sum().sort_values()
print(region_highest_loss.head())
```

```python
print("Customer Segment :")
customer_segment = df.groupby('Segment')['Profit'].sum().sort_values(ascending = False)
print(customer_segment.head())
```

```python
import matplotlib.pyplot as plt

# this is used to set size of the graph
plt.figure(figsize=(6,6))

# this is used to lable and create pie chart
plt.pie(customer_segment, labels=customer_segment.index,autopct='%1.1f%%',colors=['#ff9999','#66b3ff','#99ff99'])

# this is used to set plot title
plt.title('Profit by Customer Segment')

# this is used to show the chart
plt.show()
```

```python
print("Which Category is making best sales and worst sales ?")
df.groupby('Category')['Profit'].mean()
plt.title("BEST VS WORST SALES IN CATEGORY")
sns.barplot(x='Category', y='Profit', data=df)
```

```python
pd.crosstab(df['Category'], df['Region'])
```

```python
print("Time Series Analysis: ")
print("Are our sales increasing or decreasing over time?")
df['Order Date'].dtype

print("Converting Order date from string to Date data type")
df['Order Date'] = pd.to_datetime(df['Order Date'])
df['Order Date'].dtype

monthly_sales = df.set_index('Order Date').resample('M')['Sales'].sum()

plt.figure(figsize=(12,6))
sns.lineplot(data = monthly_sales, marker = 'o', color ='purple')
plt.title("Monthly Sales Trend (Time Series)")
plt.xlabel('Date')
plt.ylabel('Total Sales')
plt.grid(True)
plt.show()
```

```python
from matplotlib.lines import lineStyles
print("Relationship Analysis (Scatter Plot)")
print("After giving the discount, Is the profit increasing ?")

plt.figure(figsize=(10,6))
sns.scatterplot(x='Sales',y='Profit' ,data = df, hue = 'Discount', palette = 'coolwarm')
plt.title('Sales vs Profit')
plt.axhline(0, color='black', linestyle = '--')
plt.show()
```

```python
print("Optimal Discount %")
best_discount = df.groupby('Discount')['Profit'].mean().reset_index()
print(best_discount)
sns.lineplot(x = 'Discount', y = 'Profit', data = best_discount, marker='o')
```

```python
print("Outlier Detection (Box Plot)")
print("In which category profit is consistent and where it suddenly drops ?")
sns.boxplot(x = 'Category', y = 'Profit', data = df)
plt.title("Profit Distribution in Category")
plt.ylim(-500, 500)
```

In above boxplot chart you can see median profit is close to 0 in Furniture Category and Box have a moderate size but value the negative outliners is high (-400,-500) that shows the profit is not stable in Furnitur.

When You analyze Office Supplies Category , median is in profit but very small and outliners are on both side but not very extreme. The profit of maximum orders is consistent margins are low but predictable.This is the most consistent category.

Median Profit is Highest in Technology Category.Positive Outliners are High.
Box is wide which means high variability. There is high Profit Potential but it the risks are high too.

According to my Analysis The Office Supplies Category is best because it provides consistency.
